```python
def plot_quad_fig(data, subsets, legend=True):
    xs,qerrs,flow_errs,plan_errs,pg_errs,und_plan_errs,quadratic_errs,quad_flow_errs,flow_plan_errs,_ = data[subsets]
    subsets = subsets.replace("[","")
    subsets = subsets.replace("]","")
    
    plt.figure(figsize=(5,4))
    plt.style.use('ggplot')
    ax = plt.gca()
    
    quad_flow_errs = list(zip(*quad_flow_errs))
    quad_flow_errs = np.array(quad_flow_errs)
    quad_flow_errs[quad_flow_errs < 0.0] = np.nan
    
    #cycol = cycle(colors)
    lines = []
    labels = []
    
    NUM_COLORS = 11
    colors = "kcmyb"
    colors = ["k", "violet", "c", "m", "y", "b"]
    
    print(QUAD_BETAS)
    for i, beta in enumerate(QUAD_BETAS):
        cur_ax = ax
        sns.lineplot(xs,quad_flow_errs[i],label=beta, ax=cur_ax, color=colors[i], 
                     linewidth=5.0, zorder=5.0)
        
    ax.set_xscale("log")
    ax.set_xlabel("Cardinality Estimate")
    ax.get_yaxis().set_visible(False)

    if legend:
#        print("legend")
#         ax.legend(lines, [l.get_label() for l in lines], 
#                   fontsize=10, loc='lower left',
#                    bbox_to_anchor=(0.0,1.05), ncol=3, frameon=False, borderaxespad=0)
        ax.legend(loc='lower left',
                   bbox_to_anchor=(0.0,1.05), title="Beta",
                  ncol=6, frameon=False, borderaxespad=0, fontsize=8)
        

    #plt.tight_layout()
    fn = SAVE_DIR + "Beta-Errors-" + str(subsets) + ".png"
    print(fn)
    plt.savefig(fn, bbox_inches='tight')
```

```python
print("num subsets: ", len(data))
for subset in data:
    plot_quad_fig(data, subset)
```
